```python
import deepstack.core as ds
from PIL import Image
import pprint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
IP_ADDRESS = 'localhost'
PORT = '5000'
API_KEY = ""
TIMEOUT = 20 # Default is 10
```

## Scene detection

```python
image_path = 'tests/images/test-image3.jpg'

plt.imshow(Image.open(image_path));
```

## Actual

```python
dsscene = ds.DeepstackScene(IP_ADDRESS, PORT, API_KEY, TIMEOUT)
```

Perform object detection - can be slow

```python
%%time
try:
    with open(image_path, 'rb') as image_bytes:
        dsscene.detect(image_bytes)
except ds.DeepstackException as exc:
    print(exc)
```

```python
dsscene.predictions
```
